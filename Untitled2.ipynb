```python

# Train and assess model performance
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {'MSE': mse, 'R-squared': r2}

# Print model performance
for name, metrics in results.items():
    print(f"{name}: MSE={metrics['MSE']}, R-squared={metrics['R-squared']}")
```

```python
# Hyperparameter optimization for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

print("Best parameters for Random Forest:", grid_search.best_params_)
```

```python
# Fit the grid search object to perform hyperparameter tuning
grid_search.fit(X_train, y_train)
```

```python

# Access the best estimator
best_rf = grid_search.best_estimator_
```

```python

# Now you can use the best estimator for predictions
y_pred_best = best_rf.predict(X_test)
mse_best = mean_squared_error(y_test, y_pred_best)
r2_best = r2_score(y_test, y_pred_best)
print(f"Random Forest (optimized): MSE={mse_best}, R-squared={r2_best}")
```
